```python
import matplotlib.pyplot as plt

# Define the target variable
target_variable = 'Historical_Cost_of_Ride'

# Get the list of features
features = [col for col in df.columns if col != target_variable]

# Create a figure with subplots
plt.figure(figsize=(16, 12))

# Plot each feature against the target variable
for i, feature in enumerate(features):
    plt.subplot(4, 3, i + 1)  # Adjust subplot grid size as needed
    plt.scatter(df[feature], df[target_variable], alpha=0.5)
    plt.xlabel(feature)
    plt.ylabel(target_variable)
    plt.title(f'{feature} vs {target_variable}')

# Adjust layout to prevent overlapping
plt.tight_layout()
plt.show()
```

```python

vehicle_types = df['Vehicle_Type'].unique()
colors = plt.cm.tab10(range(len(vehicle_types)))
color_mapping = dict(zip(vehicle_types, colors))

bar_colors = df['Vehicle_Type'].map(color_mapping)

plt.figure(figsize=(10, 8))
plt.barh(df['Time_of_Booking'], df['Number_of_Drivers'], color=bar_colors)

plt.xlabel('Time of Booking')
plt.ylabel('Number of Drivers')
plt.title('Time of Booking vs Number of Drivers by Vehicle Type')

handles = [plt.Rectangle((0,0),1,1, color=color_mapping[vt]) for vt in vehicle_types]
plt.legend(handles, vehicle_types, loc='upper right')
plt.show()
```
