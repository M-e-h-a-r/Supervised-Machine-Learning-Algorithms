## Let's apply decision tree classifier

```python
from sklearn.preprocessing import LabelEncoder
```

```python
job=LabelEncoder()
marital=LabelEncoder()
education=LabelEncoder()
default=LabelEncoder()
housing=LabelEncoder()
loan=LabelEncoder()
contact=LabelEncoder()
month=LabelEncoder()
poutcome=LabelEncoder()
```

```python
df['job']=job.fit_transform(df['job'])
df['marital']=marital.fit_transform(df['marital'])
df['education']=education.fit_transform(df['education'])
df['default']=default.fit_transform(df['default'])
df['housing']=housing.fit_transform(df['housing'])
df['loan']=loan.fit_transform(df['loan'])
df['contact']=contact.fit_transform(df['contact'])
df['month']=month.fit_transform(df['month'])
df['poutcome']=poutcome.fit_transform(df['poutcome'])
```

```python
df
```

```python
x=df.iloc[:,0:-1]
x
```

```python
y=df['y']
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(x,y,random_state=4,test_size=0.2)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dc1=DecisionTreeClassifier(criterion='entropy')
```

```python
dc1.fit(X_train,y_train)
```

```python
y_pred=dc1.predict(X_test)
```

```python
dc1.score(X_test,y_test)
```

```python
y_pred
```

## Let's try with gini impurtiy

```python
dc=DecisionTreeClassifier(criterion='gini')
```

```python
dc.fit(X_train,y_train)
```

```python
y1_pred=dc.predict(X_test)
y1_pred
```

```python
dc.score(X_test,y_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rc=RandomForestClassifier(n_estimators=200,criterion='gini')
```

```python
rc.fit(X_train,y_train)
```

```python
y_pred2=rc.predict(X_test)
```

```python
y_pred2
```
